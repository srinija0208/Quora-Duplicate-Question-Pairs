```python
## no. of repeated and unique questions
questions = pd.Series(data['qid1'].tolist() + data['qid2'].tolist())
print(f"Number of unique questions: {questions.unique().shape[0]}")

dup = questions.value_counts() > 1
print(f"Number of duplicate questions: {dup[dup].shape[0]}")
```

```python
# Repeated questions histogram

plt.hist(questions.value_counts().values,bins=160)
plt.yscale('log')
plt.show()
```
